Write a Python function called rentals_month that retrieves rental data for a given month and year (passed as parameters) from the Sakila database as a Pandas DataFrame. The function should take in three parameters:

engine: an object representing the database connection engine to be used to establish a connection to the Sakila database.
month: an integer representing the month for which rental data is to be retrieved.
year: an integer representing the year for which rental data is to be retrieved.
The function should execute a SQL query to retrieve the rental data for the specified month and year from the rental table in the Sakila database, and return it as a pandas DataFrame.

```python
# I have created a function, where we have to insert the string (database url) into the engine to create the connection and
# both the selected month and year into the integers. I show the entire dataset, not excluding any columns

def rentals_month(connection_string:str,month:int,year:int)->pd.DataFrame:

    '''
    This function takes the connection string to create the engine and returns the number of rental per year and month for the given inputs
    
    Parameters:
    connection_string: the path to our local database
    month: month number int
    year: year number in
    
    Returns:
    
    Number of rental for the given year and month.

    '''
    engine=create_engine(connection_string)

    query = f"""SELECT *, MONTH(rental_date) AS rental_month,YEAR(rental_date) AS rental_year FROM sakila.rental
    WHERE MONTH(rental_date)={month} 
    AND YEAR(rental_date)={year};
    """
    df = pd.read_sql_query(query, engine)
    return df
```

Develop a Python function called rental_count_month that takes the DataFrame provided by rentals_month as input along with the month and year and returns a new DataFrame containing the number of rentals made by each customer_id during the selected month and year.

The function should also include the month and year as parameters and use them to name the new column according to the month and year, for example, if the input month is 05 and the year is 2005, the column name should be "rentals_05_2005".

```python
# For this function we take a dataset obtained with the previous function rentals_month and I calculate the rental count per customer id
# I ask as input the month and year as integers and create a new string to rename the rental_id column as rentals_month_year

def rental_count_month(df:pd.DataFrame,month:int,year:int)->pd.DataFrame:

    '''
    This function takes a dataframe with rental data of a mont and returns the number or rentals made by each customer
    during the selected month and year
    Parameters:
    df: df obtained from the rentals_month function
    month: month number int
    year: year number in
    
    Returns:
    
    df with number or rentals made by each customer
    during the selected month and year
    '''
    
    column_name= "rentals_"+"0"+str(month)+"_"+str(year)
    df = pd.DataFrame(df.groupby(["customer_id"])["rental_id"].nunique()).reset_index()
    df = df.rename(columns={"rental_id": column_name})
    
    return df
```

Create a Python function called compare_rentals that takes two DataFrames as input containing the number of rentals made by each customer in different months and years. 
The function should return a combined DataFrame with a new 'difference' column, which is the difference between the number of rentals in the two months.

```python
# I create the funtion to pass 2 df with data from 2 months and year and return a df with the differences in rentals.
# Since we are looking at customers with rentals in both months I have decided to go with an inner join on the merge
# I then multiply * -1 to get a better output over the difference between month1 and month2

def compare_rentals(df1:pd.DataFrame,df2:pd.DataFrame)->pd.DataFrame:
    '''
    This function takes 2 datasets and compares the rentals from each customer between 2 time ranges month_x from year_x - month_y from year_y 
    Parameters:
    df1: first dataset for a specific period
    df2: second dataset for a specific period
    
    
    Returns:
    
    df with the difference on rental per customer_id between both time periods
    '''
    
    merged_df = pd.merge(df1, df2, on='customer_id', how='inner')
    merged_df["difference"]=(merged_df.iloc[:,1]-merged_df.iloc[:,2])*-1
    return merged_df
    
```
